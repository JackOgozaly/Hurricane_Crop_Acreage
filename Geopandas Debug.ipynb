```python
#Just used for plotting
import matplotlib.pyplot as plt 
import matplotlib.patches as mpatches

#Used for the demo
from shapely.geometry import Polygon
import geopandas

#Let us create our example hurricane shapefile and geodataframe
polys1 = geopandas.GeoSeries(Polygon([(0,0), (10,0), (10,10), (0,10)]))
df1 = geopandas.GeoDataFrame({'geometry': polys1, 'df1_data':[1]})

#Let us create our example county shapefile 
polys2 = geopandas.GeoSeries(Polygon([(4,4), (4,15), (6,15), (6,4)]))
df2 = geopandas.GeoDataFrame({'geometry': polys2, 'df2_data':[1]})
```

### What our intersecting shapes look like (Go Gators)

```python
#Let's plot what our two shapes look like
fig, ax = plt.subplots(figsize=(10,8))
df1.plot(ax=ax, color = '#0000FF', alpha=.9)
df2.plot(ax=ax, color = '#FA4616', alpha=.9)
```

### What Geopandas Overlay Is Doing

```python
fig, ax = plt.subplots(figsize=(10,8))
df1.plot(ax=ax, color = '#0000FF', alpha=.9)
df2.plot(ax=ax, color = '#FA4616', alpha=.9)

#What you asked geopandas to do was create a new shape that is the intersection of the two shapes you gave it 
overlay_test = geopandas.overlay(df1, df2, how='intersection')
overlay_test.plot(ax=ax, color = '#00FF00')
```

```python
#What you are currently doing
print(overlay_test.geometry.area)

#As a little spoiler, the green rectanble has a height of 6 and a width of 2....
```
